```python
#Rating promedio por estudio
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('movies.csv', encoding="UTF-8")
studios = pd.read_csv('studios.csv', encoding="UTF-8")

#Merge de los dos dataframes juntando por id
merged = pd.merge(movies,studios, on="id", how="inner", suffixes =('_m','_s'))
merged_filtered = merged.dropna(subset=['rating'])
merged_filtered
```

```python
#groupby para agrupar por estudio, aplicando promedio en rating
relacion_estudio_promedio_ranking = merged_filtered.groupby('studio')['rating'].mean().sort_values(ascending=False)
relacion_estudio_promedio_ranking

```
