```python
#Frecuencia de generos 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movie_genres = pd.read_csv('genres.csv', encoding="UTF-8")
movie_genres.head()
```

```python
grouped_genres = movie_genres.groupby('genre').count().reset_index()
grouped_genres = grouped_genres.rename(columns={'id': 'count'})
grouped_genres = grouped_genres.sort_values(by="count", ascending = True)
grouped_genres
```

```python
plt.bar(grouped_genres["genre"], grouped_genres["count"])
plt.xticks(rotation = 45, ha="right")
plt.show()
```
